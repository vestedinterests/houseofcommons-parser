```python
import pandas as pd
speeches = pd.read_json("speeches.jsonl", lines=True)
```

```python
import numpy as np
import seaborn
import matplotlib.pyplot as plt
plt.hist([len(speech) for speech in speeches["text"]], bins= 15)
plt.xlim([0, 50])
plt.show()
```

```python
import seaborn
seaborn.histplot([len(speech) for speech in speeches["text"]], binwidth = 5)
plt.xlim(0, 40)
```
